```python
import gmaps
import pandas as pd

from config import gkey

gmaps.configure(api_key = gkey)

chicago_pharmacy_df = pd.read_csv('pharmacy-status-1.csv')
chicago_pharmacy_df.head()
```

```python
# cleaning the data
chicago_pharmacy_df.dropna(subset=['New Georeferenced Column'], inplace=True)
chicago_pharmacy_df['Location'] = chicago_pharmacy_df['New Georeferenced Column'].str.slice(start=7)
chicago_pharmacy_df['Location'] = chicago_pharmacy_df['Location'].str.replace(")","")
chicago_pharmacy_df[["Lng", "Lat"]] = chicago_pharmacy_df['Location'].str.split(expand=True)
chicago_pharmacy_df = chicago_pharmacy_df.drop(columns=["24h store?", "If closed, reopening comments (if known)", "New Georeferenced Column"])
chicago_pharmacy_df["Zip"] = chicago_pharmacy_df["Zip"].astype(str)
chicago_pharmacy_df["Zip"]=chicago_pharmacy_df["Zip"].str.slice(stop=5)
chicago_pharmacy_df.head()
```
